# Bayesian models

```python
import importlib
import sys
if 'bayesian_analysis' in sys.modules.keys():
    del sys.modules['bayesian_analysis']
from bayesian_analysis import run_bayesian_ranking_analysis, analyze_posterior, plot_posterior_differences, validate_model_fit
```

## Time

```python
mcmc_results_time = run_bayesian_ranking_analysis(
    df=time,
    n_walkers=32,
    n_steps=5000,
    n_burn=1000,
    prior_std=2.0,
    verbose=True
)
```

```python
validation_time = validate_model_fit(mcmc_results_time, time, verbose=True)

# Test 1: Superiority
print("\n" + "="*70)
print("TEST 1: Is Benchmarker 5 SUPERIOR to Benchmarkers 6 and 7?")
print("="*70)

superiority_analysis = analyze_posterior(
    results=mcmc_results_time,
    target_benchmarker=5,
    compare_against=[6, 7],
    test_type='superiority',
    credible_level=0.95,
    verbose=True,
    validation=validation_time
)

# Test 2: Inferiority
print("\n" + "="*70)
print("TEST 2: Is Benchmarker 5 INFERIOR to Benchmarkers 1, 2, 3, and 4?")
print("="*70)

inferiority_analysis = analyze_posterior(
    results=mcmc_results_time,
    target_benchmarker=5,
    compare_against=[1, 2, 3, 4],
    test_type='inferiority',
    credible_level=0.95,
    verbose=True,
    validation=validation_time
)
```

```python
# Visualizations
superiority_time = plot_posterior_differences(
    mcmc_results_time, 
    target=5, 
    compare_against=[6, 7],
    test_type='superiority')
plt.show()

inferiority_time = plot_posterior_differences(
    mcmc_results_time,
    target=5,
    compare_against=[1, 2, 3, 4], 
    test_type='inferiority'
)
plt.show()
```
